6. CPM (critical path method) in Python

```python
def topological_sort(graph):
    n = len(graph)
    visited = [False] * n
    stack = []

    def dfs(v):
        visited[v] = True
        for i, weight in enumerate(graph[v]):
            if weight and not visited[i]:
                dfs(i)
        stack.insert(0, v)

    for i in range(n):
        if not visited[i]:
            dfs(i)

    return stack

def longest_path_to_node(graph, target_node):
    order = topological_sort(graph)
    distances = [-float('inf')] * len(graph)
    predecessor = [None] * len(graph)
    start_node = order[0]
    distances[start_node] = 0

    for node in order:
        for adj in range(len(graph[node])):
            if graph[node][adj] != 0 and distances[adj] < distances[node] + graph[node][adj]:
                distances[adj] = distances[node] + graph[node][adj]
                predecessor[adj] = node

    path = []
    current = target_node
    while current is not None:
        path.insert(0, current)
        current = predecessor[current]

    if path[0] != start_node:
        return None

    return path, distances[target_node]


critical_path, path_length = longest_path_to_node(w, 11)
if critical_path:
    print("Critical path to node 11:", critical_path)
    print("Length of the critical path:", path_length)
else:
    print("No path to node 11.")
```
